**Question 3: Explain the difference between discrete and continuous distributions.**
**Answer - 3 :**

| Feature                    | **Discrete Distribution**                                                      | **Continuous Distribution**                                                                       |
| -------------------------- | ------------------------------------------------------------------------------ | ------------------------------------------------------------------------------------------------- |
| **Definition**             | Probability distribution of a **discrete random variable** (countable values). | Probability distribution of a **continuous random variable** (uncountable values in an interval). |
| **Possible values**        | Finite or countably infinite (e.g., $0,1,2,3...$).                             | Uncountably infinite within a range (e.g., real numbers between 0 and 1).                         |
| **Probability measure**    | Defined using **Probability Mass Function (PMF)**: $P(X = x)$.                 | Defined using **Probability Density Function (PDF)**: $f(x)$.                                     |
| **Probability at a point** | $P(X = x)$ can be **non-zero**.                                                | $ P(X = x) = 0$ always; probability is given over an **interval**.                                |
| **Examples**               | Number of students in a class, number of heads in coin tosses.                 | Height of students, time taken to finish a task.                                                  |

**Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
size = 1000

data = np.random.normal(mean, std_dev, size)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std = np.std(data)

print("Calculated Mean:", calculated_mean)
print("Calculated Standard Deviation:", calculated_std)

# Draw histogram
plt.hist(data, bins=30, edgecolor='black', density=True)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
‚óè Write the Python code to compute the mean sales and its confidence interval**

The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean is approximately normal if the sample size is sufficiently large, regardless of the population distribution.

```python
import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Sample statistics
n = len(data)
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation

# Confidence interval calculation
confidence_level = 0.95
alpha = 1 - confidence_level
z_value = stats.norm.ppf(1 - alpha/2)  # Z-critical value for 95%

margin_of_error = z_value * (std_sales / np.sqrt(n))
confidence_interval = (mean_sales - margin_of_error, mean_sales + margin_of_error)

# Results
print("Sample Mean Sales:", mean_sales)
print("Sample Standard Deviation:", std_sales)
print("95% Confidence Interval:", confidence_interval)
```

Step 3: Interpretation

The output gives the sample mean (our best estimate of the average daily sales).

The 95% CI provides a range where the true population mean is likely to lie.

This interval helps the company understand the overall sales trend with statistical confidence.
